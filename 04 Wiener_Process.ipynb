```python
import numpy.random as npr, numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
# The Wiener process is called brownian motion
def wiener_process(dt= 0.01, x0= 0, n= 1000):

  #W(t=0)=0
  # initial W(t) with zero
  W= np.zeros(n+1)

  # Create N+1 timesteps: t=0,1,2,3,4,...,N
  t= np.linspace(x0, n, n+1)

  # Use cumulative sum: on every step the additional value is
  # drawn from a normal distribution with mean 0 and variance dt ... N(0,dt)
  # by the way, N(0,dt) = sqrt(dt)*N(0,1) ususaly this formula is used.
  W[1:n+1] = np.cumsum(np.random.normal(0, np.sqrt(dt), n))
  # W[1:n+1] = np.cumsum(np.random.normal(1, 5, n))
  # W[1:n+1] = np.cumsum(np.random.normal(1, 10, n))
  return t, W

def plot_process(t, W):
  plt.plot(t, W)
  plt.xlabel('Time(t)')
  plt.ylabel('Wiener-process W(t)')
  plt.title('Wiener-process')
  plt.show()

if __name__ == '__main__':
  time, data = wiener_process()
  plot_process(time, data)
```
